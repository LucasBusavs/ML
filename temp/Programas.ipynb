```python
import h2o
```

```python
h2o.init()
```

```python
from h2o.automl import H2OAutoML
```

```python
airLines_df = h2o.import_file('C:/Users/Edgar/Downloads/airlines_all.05p.csv')
```

```python
airLines_df.types
```

```python
airLines_df.describe()
```

```python
airLines_TrainSet,airLines_TestSet,airLines_ValidationSet = airLines_df.split_frame(ratios=[.7,.15])
```

```python
airLines_TrainSet
```

```python
y="Cancelled"
x=airLines_df.columns
x.remove(y)
```

```python
aml = H2OAutoML(max_models = 20, seed = 10, exclude_algos = ["StackedEnsemble","DeepLearning"],verbosity='info',nfolds=0)
```

```python
aml.train(x=x, y=y ,training_frame = airLines_TrainSet, validation_frame = airLines_ValidationSet)
```

```python
MelhoresModelos = aml.leaderboard
```

```python
MelhoresModelos
```

```python
airlines_Pred = aml.leader.predict(airLines_TestSet)
```
